```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, root_mean_squared_error
```

```python
df = pd.read_csv('Student_Performance.csv')
```

```python
df
```

```python
df.corr(numeric_only=True)
```

```python
df.duplicated().sum()
```

```python
df  = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df['Extracurricular Activities'] = df['Extracurricular Activities'].map({'Yes' : 1, 'No' : 0})
```

```python
df.columns
```

```python
# if we have any one column has strong relation with output column then we can use linear regression
df.corr(numeric_only=True)
```

```python
x = df[['Hours Studied', 'Previous Scores', 'Extracurricular Activities', 'Sleep Hours', 'Sample Question Papers Practiced']]
y = df['Performance Index']
```

```python
# train_test_split() - returns 4 values
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size=0.8, random_state=42)
```

```python
print(xtrain)
```

```python
# model creation
model = LinearRegression()
```

```python
# model training
# fit() used for model training it learns the patterns from input and output
model.fit(xtrain, ytrain)
```

```python
model.intercept_
```

Intercept conclusion

Base or initial value of 'performance index' is -33.981324496440706 when all features equals to zero

```python
model.coef_
```
